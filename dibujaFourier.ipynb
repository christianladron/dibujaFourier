```julia
t = 0:3e-2:2pi
y = zeros(length(t),length(t));
```

$$u(x,y) = \big[\sum_{i=1}^3 a_i \sin(nx)\big]\big[\sum_{i=1}^3 b_i \sin(ny)\big]$$
$$x,y \in [0,2\pi]$$

```julia
lafig = figure()
@manipulate for a1=-1:0.2:1, a2=-1:0.2:1, a3=-1:0.2:1, b1=-1:0.2:1, b2=-1:0.2:1, b3=-1:0.2:1
    withfig(lafig) do
        largo = 1:length(t)
        zlim(-1,1)
        for i=largo, j=largo
            y[i,j] = (a1*sin(t[i]) + a2*sin(2*t[i])+ a3*sin(3*t[i])) * (b1*sin(t[j]) + b2*sin(2*t[j])+ b3*sin(3*t[j]))
        end
        surf(t,t,y)
    end
end
```
